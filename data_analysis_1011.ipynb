## 5.2.6 Function Application and Mapping 函数应用和映射
#### NumPy函数可以用

```python
import numpy as np
import pandas as pd  ## 默认导入模式
from pandas import Series,DataFrame  ## 常用的两个工具 
frame=pd.DataFrame(np.random.randn(4,3),columns=list('bde'),
                  index=['Utah','Ohio','Texas','Oregon'])
frame
```

```python
np.abs(frame)  # 取绝对值
```

#### 如果想要实现，一列中的最大值-最小值
#### 也就是要对每一列执行同一个函数操作
### 用apply

```python
f=lambda x:x.max()-x.min()
frame.apply(f)
```

```python
frame.apply(f,axis='columns')
#  沿着columns的方向
#  每行来一次
```

#### 返回的值可以带有多个值的Series,比如想留下最大值和最小值

```python
def f(x):
    return pd.Series([x.min(),x.max()],index=['min','max'])  # 每一列要做的操作
frame.apply(f)
```

#### frame里面都是浮点数
#### 这个浮点数保留两位，变成字符串
### 用applymap

```python
format=lambda x:'%.2f' % x
#  保留两位小数
frame.applymap(format)
#  这里的applymap，对每一列都进行操作
#  区别于map,是对某一列，也就是某Series进行操作
```

```python
frame['e'].map(format)
#  map方法
```

## 5.2.7 Sorting and Ranking 排序和排名
### 按索引排序，用sort_index
### 按值排序，用sort_values

```python
obj=pd.Series(range(4),index=['d','a','b','c'])
obj.sort_index()
```

```python
# 默认按axis=0排序
frame=pd.DataFrame(np.arange(8).reshape((2,4)),
                  index=['three','one'],
                  columns=['d','a','b','c'])
frame.sort_index()
```

```python
# 给axis=1排序
frame.sort_index(axis=1)
```

```python
# 默认升序，降序为ascending=False
frame.sort_index(axis=1,ascending=False)
```

```python
obj=pd.Series([4,7,-3,2])
# 默认index为0123...
obj.sort_values()
```

#### 缺失值会被排序至Series的尾部

```python
obj=pd.Series([4,np.nan,7,np.nan,-3,2])
obj.sort_values()
```

#### 给DataFrame,就是表格排序
#### sort_values(by=排序的根据)

```python
frame=pd.DataFrame({'b':[4,7,-3,2],'a':[0,1,0,1]})
frame
```

```python
frame.sort_values(by='b')
```

```python
frame.sort_values(by=['a','b'])
# 排序优先次序，a然后b
```

```python
obj=pd.Series([7,-5,7,4,2,0,4])
obj
```

```python
obj.rank()
# 按值从小到大
# 排名，1：-5排在第一位------1.0
#  5：0排在第二位------2.0
# 4：2排在第三位-----3.0
#  6：4，3：4相等，并列第四，也就是第四和第五，所以都是4.5位
# 以此类推
```

```python
obj.rank(method='first')
# 3和6对应的值都是4
# method=first 就是先看到的优先
# 也就是3：4排在第四位
# 6：4排在第五位
```

```python
obj.rank(ascending=False,method='max')
# 首先是反序，由大到小
# [7,-5,7,4,2,0,4]
# 0:7和2：7，最大，在第一位和第二位
# method=max就是取最大的值
# 也就是两个排名都是2
```

```python
frame=pd.DataFrame({'b':[4.3,7,-3,2],'a':[0,1,0,1],'c':[-2,5,8,-2.5]})
frame
```

```python
frame.rank(axis='columns')
# 明显是abc对应的三个数的排名
```

## 5.3 Summarizing and Computing Descriptive Statistics 描述性统计的概述与计算
#### 主要是进行统计运算时，对缺失值的处理

```python
df=pd.DataFrame([[1.4,np.nan],[7.1,-4.5],
                [np.nan,np.nan],[0.75,-1.3]],
                index=['a','b','c','d'],
                columns=['one','two'])
#  建立一个有缺失值的表
df
```

```python
df.sum()
# NaN默认为0
# 默认计算axis=0方向，竖着加
```

```python
df.sum(axis='columns')
# 计算axis=1方向，横着加，c行印证了NaN默认为0
```

```python
# 如果想保留缺失值，则加入参数skipna=False
df.mean(axis='columns',skipna=False)
```

```python
# 最大值的索引值，one这一列，最大值为7.10，对应索引为b
df.idxmax()
```

```python
df.cumsum()
# 积累型计算
# 1.4
# 1.4+7.1=8.5
# NaN
# 1.4+7.1+0.75=9.25
```

```python
# 一次性产生多个汇总统计
df.describe()
```

```python
# 对于非数值型数据，产生另一种汇总统计
obj=pd.Series(['a','a','b','c']*4)
obj
```

```python
obj.describe()
#  共16个元素
#  元素abc三个
#  top最多的是a
#  frequency,重复的次数是8
```

## 5.3.1 Correlation and Covariance 相关性和协方差

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame  ## 常用的两个工具 
import pandas_datareader.data as web
all_data={ticker:web.get_data_yahoo(ticker)
         for ticker in ['AAPL','IBM','MSFT','GOOG']}
# 下载股票代号ticker为这几个的数据
```

#### 取出四只股票代码和价格，通过字典生成一个DataFrame
#### 其中字典的key值来自all_data里面key的值
#### value值是all_data里面DataFrame的一列
#### 这里用的是Adj Close

```python
price=pd.DataFrame({ticker:data['Adj Close']
                   for ticker,data in all_data.items()})
price
```

```python
## 取出四只股票当日成交量
volume=pd.DataFrame({ticker:data['Volume']
                   for ticker,data in all_data.items()})
volume
```

#### 计算股票变动百分比

```python
returns=price.pct_change()
returns.tail()
```

#### 两列的相关性

```python
returns['MSFT'].corr(returns['IBM'])
# 或者returns.MSFT.corr(returns.IBM)
```

#### 两列的协方差

```python
returns['MSFT'].cov(returns['IBM'])
```

#### 返回四只股票价格变动之间的相关性矩阵

```python
returns.corr()
```

#### 返回四只股票的协方差矩阵

```python
returns.cov()
```

#### 想要每只股票与IBM股价的相关性

```python
returns.corrwith(returns.IBM)
```

#### 成交量与价格变化百分比的相关性

```python
returns.corrwith(volume)
```

#### High当日最高价
#### Low当日最低价
#### Open当日开盘价
#### Close当日收盘价
#### Volume当日成交量
#### Adj Close 前复权收盘价，为了更准确反应股票的价值，针对分红影响进行的调整

```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame  ## 常用的两个工具 
import pandas_datareader.data as web
to_match=pd.Series(['c','a','b','b','c','a','e'])
to_match
```

```python
unique_vals=pd.Series(['c','b','a'])
unique_vals
```

### Index.get_indexer方法，可以提供一个索引数组，这个索引数组可以将可能非唯一值数组转换成另一个唯一的数组

```python
pd.Index(unique_vals).get_indexer(to_match)
```

```python
l=list(unique_vals)
for x in to_match:
    if x in l:
        print(l.index(x))
    else:
        print(-1)
```

###  0 to_match[0]为c在unique_vals索引为0

###  1 to_match[0]为b在unique_vals索引为1
###  1 to_match[0]为b在unique_vals索引为1
###  0 to_match[0]为c在unique_vals索引为0
